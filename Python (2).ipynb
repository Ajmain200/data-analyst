Problem for pass ,continue ,break


Print a multiplication table from 1 to 5

Skip printing multiples of 3

Stop the inner loop if the product > 12


```python
for i in range(1, 6):
    for j in range(1, 6):
        product = i * j

        if product % 3 == 0:
            continue

        if product > 12:
            break

        if product == 10:
            pass

        print(i,"*",j,"=",product)
```

Dictionary all different uses


```python
# Step 0: Initial dictionary
students = {
    "S1": {"name": "Simon", "age": 22, "grade": "A"},
    "S2": {"name": "Rahim", "age": 21, "grade": "B"},
    "S3": {"name": "Karim", "age": 23, "grade": "C"}
}

# 1. Keys, Values, Items
print("Keys:", students.keys())
print("Values:", students.values())
print("Items:", students.items())

# 2. Get age of S2
print("S2 Age:", students.get("S2")["age"])

# 3. Add new student S4
students["S4"] = {"name": "Salma", "age": 20, "grade": "A+"}
print("After adding S4:", students)

# 4. Update grade of S3
students["S3"]["grade"] = "B+"
print("After updating S3 grade:", students)

# 5. Remove student S1
students.pop("S1")
print("After removing S1:", students)

# 6. Remove last inserted item
students.popitem()
print("After popitem():", students)

# 7. Clear dictionary
students.clear()
print("After clearing:", students)
```

```python
#functions problem,find greater from 2 values
def find_greater(a, b):
    if a > b:
        return a
    else:
        return b
find_greater(5,6)
```

```python
#recursion

def fact(n):
    if n==0 or n==1:
        return 1
    return n * fact(n-1)

print(fact(5))
```

```python
def fib(n):
    if n<=1:
        return n
    return fib(n-1) + fib(n-2)

print(fib(6))
```

Class University problem


```python
class Person:
    count_people = 0  # class variable

    def __init__(self, name, age):
        self.name = name        # instance variable
        self.age = age
        Person.count_people += 1

    def show_info(self):
        print(f"Person: {self.name}, Age={self.age}")

    @staticmethod
    def is_adult(age):
        return age >= 18

    @classmethod
    def total_people(cls):
        return cls.count_people


class Student(Person):
    def __init__(self, name, age, dept):
        super().__init__(name, age)
        self.dept = dept

    def show_info(self):
        print(f"Student: {self.name}, Age={self.age}, Dept={self.dept}")


class Employee(Person):
    def __init__(self, name, age, salary):
        super().__init__(name, age)
        self.__salary = salary  # private variable

    def get_salary(self):
        return self.__salary

    def show_info(self):
        print(f"Employee: {self.name}, Age={self.age}, Salary={self.__salary}")


class TA(Student, Employee):
    def __init__(self, name, age, dept, salary):
        # initialize Person only once
        Person.__init__(self, name, age)
        self.dept = dept
        self.__salary = salary

    def show_info(self):
        print(f"TA: {self.name}, Age={self.age}, Dept={self.dept}, Salary={self.__salary}")


# create objects
p = Person("Rahim", 40)
s = Student("Karim", 22, "CSE")
e = Employee("Salma", 30, 30000)
t = TA("Babul", 25, "EEE", 15000)

# polymorphism
for obj in [p, s, e, t]:
    obj.show_info()

print("Total People =", Person.total_people())
print("Is 17 adult? ->", Person.is_adult(17))
```
