#Steps to install Python and use it in VS Code.

Step-1: Install Python

Go to: https://www.python.org/downloads/

Download latest version

Run the installer

VERY IMPORTANT: check the box “Add Python to PATH”

Click Install

Step-2: Install Visual Studio Code

Go to: https://code.visualstudio.com/

Download VS Code and install

Step-3: Install Python Extension in VS Code

Open VS Code

Go to Extensions (left side icon or press Ctrl + Shift + X)

Search: Python

Install the "Python" extension by Microsoft

Step-4: Create and Run Python file

Create a folder (example: python_projects)

Open that folder in VS Code

Create a file: test.py

Write:

print("Hello Python")


Run the file:

Right click Run Python File in Terminal
OR

click the “Run” button above the code

You will see the output in the terminal.

Step-5: Select Python Interpreter (if needed)

Sometimes VS Code asks for interpreter.

Press Ctrl + Shift + P

Type: Python: Select Interpreter

Choose the Python version you installed
